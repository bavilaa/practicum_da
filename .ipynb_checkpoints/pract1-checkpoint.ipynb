```python
import pandas as pd
%config IPCompleter.greedy=True

```

```python
#loading dataset
df = pd.read_csv('Consumer_Complaints.csv')
```

```python
#Checking data types
df.dtypes
```

```python
#Sample dataset

df2 = df.sample(1000)


#df2 = df

```

```python
#Transforming date into date object

df2['new_date'] = pd.to_datetime(df2['Date received'], 
                                 format='%m/%d/%Y', 
                                 errors='coerce').apply(lambda dt: dt.replace(day=1)).dt.date
df2['new_date'] = pd.to_datetime(df2['new_date'])

#Sorting

df2 = df2.sort_values(by=['new_date'],ascending=False)

#Extracting year, month, day_name
df2['year'] = pd.DatetimeIndex(df2['Date received']).year
df2['month'] = pd.DatetimeIndex(df2['Date received']).month
df2['day'] = pd.DatetimeIndex(df2['Date received']).day_name()
```

```python
df2.head()
```

```python
#Unique elements:

df_unique = df2.drop_duplicates(subset=['Product'])
```
