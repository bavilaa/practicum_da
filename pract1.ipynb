```python
%config IPCompleter.greedy=True

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set()
```

```python
#loading dataset

bd_complanits = "Consumer_Complaints.csv"

df = pd.read_csv('Consumer_Complaints.csv')
```

```python
#Checking data types
df.dtypes
```

```python
#Sample dataset
df2 = df.sample(10000)
#df2 = df

```

```python
#Transforming date into date object
df2['new_date'] = pd.to_datetime(df2['Date received'], 
                                 format='%m/%d/%Y', 
                                 errors='coerce')
df2['new_date'] = pd.to_datetime(df2['new_date'])

#Sorting

df2 = df2.sort_values(by=['new_date'],ascending=False)

#Extracting year, month, day_name
df2['year'] = pd.DatetimeIndex(df2['new_date']).year
df2['month'] = pd.DatetimeIndex(df2['new_date']).month
df2['day'] = pd.DatetimeIndex(df2['new_date']).day_name()
df2['day_number'] = pd.DatetimeIndex(df2['new_date']).day
```

```python
df2.head()
```

```python
#Unique elements:

df_unique = df2.drop_duplicates(subset=['Product'])
```

```python
#df_unique
```

```python
#Counting the number of complaints by date and product:

#df2_ncomplaints = df2.groupby(['Product','Date received']).count()


df2.columns

df2_ncomplaints = df2[['Date received' , 'new_date' , 'Product' , 'year']]


df2_ncomplaints = df2_ncomplaints.groupby(['Product' , 'new_date','year'] , as_index=False).count()


#long to wide format

df2_ncomplaints_t = pd.pivot_table(df2_ncomplaints, values='Date received', index=['new_date', 'year'],
                   columns=['Product'], aggfunc=np.sum)



#.sort_values(by=['Date received'],ascending=False)

#[['Product']]
#df2_ncomplaints_t = pd.pivot_table(df2, values='D', index=['A', 'B'],
#                   columns=['C'], aggfunc=np.sum)
```

```python
df2_ncomplaints_t = df2_ncomplaints_t.reset_index()
```

```python
#Plotting some results:
fig, ax = plt.subplots(figsize=(12, 4))
df2_ncomplaints_t = df2_ncomplaints_t[df2_ncomplaints_t.year>=2015]





sns.lineplot(x="new_date", y="Mortgage",
legend="full", data=df2_ncomplaints_t);



#sns.lineplot(data=df2_ncomplaints_t.Mortgage, color="b", ax=ax)





```
